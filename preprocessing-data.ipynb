```python
import os
from fnmatch import fnmatch
import pandas as pd
import cv2

def convert_bytes(num):
    """
    this function will convert bytes to MB.... GB... etc
    """
    for x in ['bytes', 'KB', 'MB', 'GB', 'TB']:
        if num < 1024.0:
            return "%3.1f %s" % (num, x)
        num /= 1024.0


def file_size(file_path):
    """
    this function will return the file size
    """
    if os.path.isfile(file_path):
        file_info = os.stat(file_path)
        #return convert_bytes(file_info.st_size)
        return file_info.st_size
```

```python
root = '/home/mayank/Desktop/gender-bias-detection/official-trailers/'
pattern = "*.mp4"

columns = ['path', 'filename','movie-name', 'year-of-release', 'size', 'trailer', 'movie']

df = pd.DataFrame(columns=columns)

for path, subdirs, files in os.walk(root):
    for name in files:
        if fnmatch(name, pattern):
            
            # filename = os.path.join(path, name)
            # print('filename', filename)
            # print('filename type', type(filename))
            # print('path', path)
            # print('path type', type(path))
            # print('files', files)
            # print('files type', type(files))
            # print('files[0]', files[0])
            # break
            
            p = os.path.join(path, name)
            filename = files[0]
            year = filename.split('-')[0]
            movie_name = filename.split('-')[1]
            movie_name = movie_name.split('.')[0]
            size = file_size(p)
            if (size < 5e+7) :
                trailer = 1
                movie = 0
            else :
                trailer = 0
                movie = 1
                 
            df.loc[df.shape[0]] = [p, filename, movie_name, year, size, trailer, movie] 
           
            
df.head()

df.to_csv('/home/mayank/Desktop/gender-bias-detection/trailers_info.csv')
```
